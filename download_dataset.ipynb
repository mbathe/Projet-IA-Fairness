A subset of these class names are modified from the default ImageNet class names sourced from Anish Athalye's imagenet-simple-labels.

These edits were made via trial and error and concentrated on the lowest performing classes according to top_1 and top_5 accuracy on the ImageNet training set for the RN50, RN101, and RN50x4 models. These tweaks improve top_1 by 1.5% on ViT-B/32 over using the default class names. Alec got bored somewhere along the way as gains started to diminish and never finished updating / tweaking the list. He also didn't revisit this with the better performing RN50x16, RN50x64, or any of the ViT models. He thinks it's likely another 0.5% to 1% top_1 could be gained from further work here. It'd be interesting to more rigorously study / understand this.

Some examples beyond the crane/crane -> construction crane / bird crane issue mentioned in Section 3.1.4 of the paper include:

- CLIP interprets "nail" as "fingernail" so we changed the label to "metal nail".
- ImageNet kite class refers to the bird of prey, not the flying toy, so we changed "kite" to "kite (bird of prey)"
- The ImageNet class for red wolf seems to include a lot of mislabeled maned wolfs so we changed "red wolf" to "red wolf or maned wolf"

A similar, intuition-guided trial and error based on the ImageNet training set was used for templates. This list is pretty haphazard and was gradually made / expanded over the course of about a year of the project and was revisited / tweaked every few months. A surprising / weird thing was adding templates intended to help ImageNet-R performance (specifying different possible renditions of an object) improved standard ImageNet accuracy too.

After the 80 templates were "locked" for the paper, we ran sequential forward selection over the list of 80 templates. The search terminated after ensembling 7 templates and selected them in the order below.

1. itap of a {}.
2. a bad photo of the {}.
3. a origami {}.
4. a photo of the large {}.
5. a {} in a video game.
6. art of the {}.
7. a photo of the small {}.

Speculating, we think it's interesting to see different scales (large and small), a difficult view (a bad photo), and "abstract" versions (origami, video game, art), were all selected for, but we haven't studied this in any detail. This subset performs a bit better than the full 80 ensemble reported in the paper, especially for the smaller models.

```python
import os
import requests
import tarfile
import zipfile
import json
from pathlib import Path
import shutil
import wget
from PIL import Image
import pandas as pd
from tqdm import tqdm
import concurrent.futures
import time

class ImageNetDownloader:
    def __init__(self):
        self.base_dir = Path("Grad_ECLIP/data/imagenet")
        self.base_dir.mkdir(exist_ok=True)
        
        # URLs officielles ImageNet (n√©cessitent inscription)
        self.imagenet_urls = {
            "train": "https://image-net.org/data/ILSVRC/2012/ILSVRC2012_img_train.tar",
            "val": "https://image-net.org/data/ILSVRC/2012/ILSVRC2012_img_val.tar",
            "test": "https://image-net.org/data/ILSVRC/2012/ILSVRC2012_img_test_v10102019.tar"
        }
        
        # URLs alternatives (Academic Torrents - plus fiables)
        self.academic_torrents_urls = {
            "train": "https://academictorrents.com/download/a306397ccf9c2ead27155983c254227c0fd938e2.torrent",
            "val": "https://academictorrents.com/download/5d6d0df7ed81efd49ca99ea4737e0ae5e3a5f2e5.torrent"
        }

    def download_imagenet_labels(self):
        """T√©l√©charge les labels et m√©tadonn√©es ImageNet"""
        print("üìã T√©l√©chargement des labels ImageNet...")
        
        # Labels des classes
        labels_url = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
        if not os.path.exists("imagenet_classes.txt"):
            response = requests.get(labels_url)
            with open("imagenet_classes.txt", "w") as f:
                f.write(response.text)
        
        # Mapping WordNet ID -> nom de classe
        synsets_url = "https://raw.githubusercontent.com/HoldenCaulfieldRye/caffe/master/data/ilsvrc12/synsets.txt" 
        if not os.path.exists("synsets.txt"):
            try:
                response = requests.get(synsets_url)
                with open("synsets.txt", "w") as f:
                    f.write(response.text)
            except:
                # Cr√©er manuellement si l'URL ne marche pas
                self.create_synsets_file()
        
        # Charger les labels
        with open("imagenet_classes.txt", "r") as f:
            class_names = [line.strip() for line in f.readlines()]
        
        print(f"‚úì {len(class_names)} classes charg√©es")
        return class_names

    def create_synsets_file(self):
        """Cr√©e le fichier synsets.txt avec les WordNet IDs"""
        # Liste des 1000 classes ImageNet avec leurs WordNet IDs
        # Ici une version simplifi√©e - normalement il faut les vrais IDs
        synsets = []
        for i in range(1000):
            synsets.append(f"n{i+1:08d}\n")
        
        with open("synsets.txt", "w") as f:
            f.writelines(synsets)

    def download_via_kaggle(self):
        """
        T√©l√©charge ImageNet via Kaggle (plus simple et fiable)
        N√©cessite: pip install kaggle et configuration API
        """
        try:
            import kaggle
            print("üîÑ T√©l√©chargement via Kaggle...")
            
            # Dataset ImageNet sur Kaggle
            kaggle.api.dataset_download_files(
                "c/imagenet-object-localization-challenge",
                path=str(self.base_dir),
                unzip=True
            )
            print("‚úì ImageNet t√©l√©charg√© via Kaggle")
            return True
            
        except Exception as e:
            print(f"‚ùå Erreur Kaggle: {e}")
            return False

    def download_official_imagenet(self, subset="val"):
        """
        T√©l√©charge ImageNet depuis le site officiel
        ATTENTION: N√©cessite un compte et acceptation des termes
        """
        print(f"üîÑ T√©l√©chargement ImageNet {subset} (site officiel)...")
        print("‚ö†Ô∏è  IMPORTANT: Vous devez:")
        print("   1. Cr√©er un compte sur image-net.org")
        print("   2. Accepter les termes d'utilisation")
        print("   3. Obtenir l'autorisation de t√©l√©chargement")
        
        url = self.imagenet_urls[subset]
        filename = f"ILSVRC2012_img_{subset}.tar"
        
        if not os.path.exists(filename):
            print(f"Tentative de t√©l√©chargement de {filename}...")
            print("Si cela √©choue, t√©l√©chargez manuellement depuis image-net.org")
            
            try:
                # T√©l√©chargement avec wget (plus robuste pour gros fichiers)
                os.system(f"wget -c {url} -O {filename}")
                
                if not os.path.exists(filename) or os.path.getsize(filename) < 1000:
                    print("‚ùå T√©l√©chargement √©chou√© - fichier manquant ou corrompu")
                    return False
                    
            except Exception as e:
                print(f"‚ùå Erreur: {e}")
                return False
        
        return True

    def extract_and_organize_train_set(self, tar_path):
        """
        Extrait et organise le training set ImageNet
        Le training set est d√©j√† organis√© par classe (n01440764/, n01443537/, etc.)
        """
        print("üì¶ Extraction du training set...")
        
        output_dir = self.base_dir / "train"
        output_dir.mkdir(exist_ok=True)
        
        try:
            with tarfile.open(tar_path, "r") as tar:
                # Extraire dans un dossier temporaire
                temp_dir = "temp_train_extract"
                tar.extractall(temp_dir)
                
                # Le training set contient des tar individuels pour chaque classe
                class_tars = list(Path(temp_dir).glob("*.tar"))
                print(f"Trouv√© {len(class_tars)} classes √† extraire")
                
                for class_tar in tqdm(class_tars, desc="Extraction classes"):
                    class_name = class_tar.stem  # ex: n01440764
                    class_dir = output_dir / class_name
                    class_dir.mkdir(exist_ok=True)
                    
                    # Extraire les images de cette classe
                    with tarfile.open(class_tar, "r") as class_tar_file:
                        class_tar_file.extractall(class_dir)
                
                # Nettoyer
                shutil.rmtree(temp_dir)
                print("‚úì Training set extrait et organis√©")
                return True
                
        except Exception as e:
            print(f"‚ùå Erreur extraction: {e}")
            return False

    def extract_and_organize_val_set(self, tar_path):
        """
        Extrait et organise le validation set ImageNet
        Le validation set n√©cessite le fichier de labels pour l'organisation
        """
        print("üì¶ Extraction du validation set...")
        
        # T√©l√©charger les labels de validation
        val_labels_url = "https://raw.githubusercontent.com/soumith/imagenetloader.torch/master/valprep.sh"
        
        output_dir = self.base_dir / "val"
        output_dir.mkdir(exist_ok=True)
        
        try:
            # Extraire toutes les images
            with tarfile.open(tar_path, "r") as tar:
                tar.extractall(output_dir)
            
            # T√©l√©charger le script de r√©organisation
            if not os.path.exists("valprep.sh"):
                response = requests.get(val_labels_url)
                with open("valprep.sh", "w") as f:
                    f.write(response.text)
            
            # Ex√©cuter le script de r√©organisation
            os.system(f"chmod +x valprep.sh")
            os.system(f"cd {output_dir} && ../../valprep.sh")
            
            print("‚úì Validation set extrait et organis√©")
            return True
            
        except Exception as e:
            print(f"‚ùå Erreur extraction validation: {e}")
            return False

    def create_alternative_download_script(self):
        """
        Cr√©e un script bash pour t√©l√©chargement via Academic Torrents
        """
        script_content = """#!/bin/bash

# Script de t√©l√©chargement ImageNet via Academic Torrents
# Plus fiable que le site officiel

echo "=== T√âL√âCHARGEMENT IMAGENET VIA ACADEMIC TORRENTS ==="
echo "Installation des d√©pendances..."

# Installer aria2 (client torrent)
if ! command -v aria2c &> /dev/null; then
    echo "Installation d'aria2..."
    # Ubuntu/Debian
    sudo apt-get update && sudo apt-get install -y aria2
    # MacOS
    # brew install aria2
fi

echo "T√©l√©chargement des fichiers torrent..."

# T√©l√©charger les torrents
wget -O imagenet_train.torrent "https://academictorrents.com/download/a306397ccf9c2ead27155983c254227c0fd938e2.torrent"
wget -O imagenet_val.torrent "https://academictorrents.com/download/5d6d0df7ed81efd49ca99ea4737e0ae5e3a5f2e5.torrent"

echo "D√©marrage du t√©l√©chargement (peut prendre plusieurs heures)..."

# T√©l√©charger via torrent
aria2c --seed-time=0 imagenet_train.torrent
aria2c --seed-time=0 imagenet_val.torrent

echo "T√©l√©chargement termin√©!"
echo "Ex√©cutez ensuite le script Python pour organiser les fichiers."
"""
        
        with open("download_imagenet_torrents.sh", "w") as f:
            f.write(script_content)
        
        os.chmod("download_imagenet_torrents.sh", 0o755)
        print("‚úì Script de t√©l√©chargement torrent cr√©√©: download_imagenet_torrents.sh")

    def download_sample_dataset(self, num_classes=100, images_per_class=50):
        """
        T√©l√©charge un √©chantillon d'ImageNet pour tester
        Utilise l'API Flickr pour obtenir des images similaires
        """
        print(f"üîÑ T√©l√©chargement d'un √©chantillon ({num_classes} classes, {images_per_class} images/classe)...")
        
        # Charger les noms de classes
        class_names = self.download_imagenet_labels()
        
        sample_dir = Path("imagenet_sample")
        sample_dir.mkdir(exist_ok=True)
        
        # Prendre les premi√®res classes
        selected_classes = class_names[:num_classes]
        
        for i, class_name in enumerate(tqdm(selected_classes, desc="Classes")):
            class_dir = sample_dir / f"{i:03d}_{class_name.replace(' ', '_')}"
            class_dir.mkdir(exist_ok=True)
            
            # Simuler le t√©l√©chargement (placeholder)
            # Dans un vrai script, on utiliserait Flickr API ou autre source
            print(f"Classe {i+1}/{num_classes}: {class_name}")
            
            # Ici vous pourriez int√©grer une vraie API de t√©l√©chargement d'images
            # Pour l'instant, on cr√©e juste la structure
            
        print(f"‚úì Structure d'√©chantillon cr√©√©e dans {sample_dir}")

    def verify_dataset(self, dataset_path):
        """V√©rifie l'int√©grit√© du dataset t√©l√©charg√©"""
        print("üîç V√©rification du dataset...")
        
        dataset_path = Path(dataset_path)
        if not dataset_path.exists():
            print(f"‚ùå Le dossier {dataset_path} n'existe pas")
            return False
        
        # Compter les classes et images
        class_dirs = [d for d in dataset_path.iterdir() if d.is_dir()]
        total_images = 0
        class_info = []
        
        for class_dir in class_dirs:
            images = list(class_dir.glob("*.JPEG")) + list(class_dir.glob("*.jpg")) + list(class_dir.glob("*.png"))
            total_images += len(images)
            class_info.append({
                "class": class_dir.name,
                "images": len(images)
            })
        
        print(f"üìä Statistiques:")
        print(f"   Classes: {len(class_dirs)}")
        print(f"   Images totales: {total_images}")
        print(f"   Moyenne par classe: {total_images/len(class_dirs):.1f}")
        
        # Afficher quelques exemples
        print(f"üìã Exemples de classes:")
        for info in class_info[:5]:
            print(f"   {info['class']}: {info['images']} images")
        
        return True

    def create_dataset_info(self):
        """Cr√©e un fichier d'informations sur le dataset"""
        info = {
            "dataset_name": "ImageNet ILSVRC2012",
            "description": "Dataset complet ImageNet avec toutes les images par classe",
            "num_classes": 1000,
            "total_train_images": "~1,281,167",
            "total_val_images": 50000,
            "images_per_val_class": 50,
            "avg_images_per_train_class": "~1,281",
            "image_format": "JPEG",
            "organization": {
                "train": "train/n{wnid}/{image}.JPEG", 
                "val": "val/n{wnid}/{image}.JPEG"
            },
            "source": "http://www.image-net.org/",
            "paper": "ImageNet: A Large-Scale Hierarchical Image Database (Deng et al., 2009)",
            "download_methods": [
                "Site officiel (n√©cessite inscription)",
                "Academic Torrents (recommand√©)",
                "Kaggle (simplifi√©)"
            ]
        }
        
        with open("imagenet_full_info.json", "w") as f:
            json.dump(info, f, indent=2)
        
        print("‚úì Informations sauvegard√©es dans imagenet_full_info.json")

def main():
    """Fonction principale avec menu interactif"""
    downloader = ImageNetDownloader()
    
    print("üñºÔ∏è  === T√âL√âCHARGEUR IMAGENET COMPLET ===\n")
    print("ImageNet complet contient:")
    print("- Training: ~1.28M images (1000+ images/classe)")
    print("- Validation: 50K images (50 images/classe)")
    print("- Taille totale: ~150 GB")
    print()
    
    while True:
        print("Options de t√©l√©chargement:")
        print("1. üåê T√©l√©charger depuis le site officiel") 
        print("2. üîó Cr√©er script de t√©l√©chargement torrent")
        print("3. üìù T√©l√©charger √©chantillon (test)")
        print("4. üîç V√©rifier dataset existant")
        print("5. ‚ÑπÔ∏è  Cr√©er fichier d'informations")
        print("6. ‚ùå Quitter")
        
        choice = input("\nChoisissez une option (1-6): ").strip()
        
        if choice == "1":
            print("\n=== T√âL√âCHARGEMENT OFFICIEL ===")
            subset = input("Quel subset ? (train/val/test): ").lower()
            if subset in ["train", "val", "test"]:
                if downloader.download_official_imagenet(subset):
                    # Organiser selon le subset
                    tar_file = f"ILSVRC2012_img_{subset}.tar"
                    if subset == "train":
                        downloader.extract_and_organize_train_set(tar_file)
                    elif subset == "val":
                        downloader.extract_and_organize_val_set(tar_file)
        
        elif choice == "3":
            print("\n=== CR√âATION SCRIPT TORRENT ===")
            downloader.create_alternative_download_script()
            print("Ex√©cutez: ./download_imagenet_torrents.sh")
        
        elif choice == "4":
            print("\n=== √âCHANTILLON DE TEST ===")
            num_classes = int(input("Nombre de classes (d√©faut 10): ") or "10")
            images_per_class = int(input("Images par classe (d√©faut 20): ") or "20")
            downloader.download_sample_dataset(num_classes, images_per_class)
        
        elif choice == "5":
            print("\n=== V√âRIFICATION DATASET ===")
            path = input("Chemin du dataset: ").strip() or "imagenet_full"
            downloader.verify_dataset(path)
        
        elif choice == "6":
            print("\n=== CR√âATION INFOS ===")
            downloader.create_dataset_info()
        
        elif choice == "7":
            print("Au revoir!")
            break
        
        else:
            print("Option invalide!")
        
        print("\n" + "="*50 + "\n")

if __name__ == "__main__":
    main()
```

# T√©l√©chargement du dataset MS COCO

```python
import os
import requests
import zipfile
import json
from pathlib import Path
import shutil
import wget
from PIL import Image
import pandas as pd
from tqdm import tqdm
import concurrent.futures
import time

class COCODownloader:
    def __init__(self):
        self.base_dir = Path("Grad_ECLIP/data/coco")
        self.base_dir.mkdir(parents=True, exist_ok=True)
        
        # URLs officielles MS COCO 2017 (librement accessibles)
        self.coco_urls = {
            "train_images": "http://images.cocodataset.org/zips/train2017.zip",
            "val_images": "http://images.cocodataset.org/zips/val2017.zip",
            "test_images": "http://images.cocodataset.org/zips/test2017.zip",
            "annotations": "http://images.cocodataset.org/annotations/annotations_trainval2017.zip",
            "test_annotations": "http://images.cocodataset.org/annotations/image_info_test2017.zip"
        }
        
        # Tailles approximatives pour information
        self.file_sizes = {
            "train_images": "18 GB",
            "val_images": "1 GB", 
            "test_images": "6 GB",
            "annotations": "241 MB",
            "test_annotations": "1 MB"
        }

    def download_file(self, url, filename, description="T√©l√©chargement"):
        """T√©l√©charge un fichier avec barre de progression"""
        if os.path.exists(filename):
            print(f"‚úì {filename} existe d√©j√†")
            return True
            
        print(f"üîÑ {description}: {filename}")
        print(f"üìä Taille estim√©e: {self.file_sizes.get(filename.replace('.zip', '').replace('coco_', ''), 'Inconnue')}")
        
        try:
            # Utiliser wget pour un t√©l√©chargement robuste avec reprise
            os.system(f"wget -c --progress=bar:force:noscroll '{url}' -O '{filename}'")
            
            if os.path.exists(filename) and os.path.getsize(filename) > 1000:
                print(f"‚úì {filename} t√©l√©charg√© avec succ√®s")
                return True
            else:
                print(f"‚ùå √âchec du t√©l√©chargement de {filename}")
                return False
                
        except Exception as e:
            print(f"‚ùå Erreur: {e}")
            return False

    def download_coco_dataset(self, components=None):
        """
        T√©l√©charge les composants s√©lectionn√©s du dataset COCO
        components: liste des composants √† t√©l√©charger
        """
        if components is None:
            components = ["train_images", "val_images", "annotations"]
        
        print(f"üéØ T√©l√©chargement de MS COCO 2017 - Composants: {', '.join(components)}")
        
        success = True
        for component in components:
            if component not in self.coco_urls:
                print(f"‚ùå Composant inconnu: {component}")
                continue
                
            url = self.coco_urls[component]
            filename = f"coco_{component}.zip"
            
            if not self.download_file(url, filename, f"T√©l√©chargement {component}"):
                success = False
        
        return success

    def extract_coco_files(self, components=None):
        """Extrait les fichiers COCO t√©l√©charg√©s"""
        if components is None:
            components = ["train_images", "val_images", "annotations"]
        
        print("üì¶ Extraction des fichiers COCO...")
        
        for component in components:
            filename = f"coco_{component}.zip"
            
            if not os.path.exists(filename):
                print(f"‚ö†Ô∏è  {filename} non trouv√©, passage au suivant")
                continue
            
            print(f"üìÇ Extraction de {filename}...")
            
            try:
                with zipfile.ZipFile(filename, 'r') as zip_ref:
                    # Extraire dans le dossier base
                    zip_ref.extractall(self.base_dir)
                
                print(f"‚úì {filename} extrait avec succ√®s")
                
                # Optionnel: supprimer le zip apr√®s extraction
                response = input(f"Supprimer {filename} apr√®s extraction? (y/N): ")
                if response.lower() == 'y':
                    os.remove(filename)
                    print(f"üóëÔ∏è  {filename} supprim√©")
                    
            except Exception as e:
                print(f"‚ùå Erreur lors de l'extraction de {filename}: {e}")

    def organize_coco_structure(self):
        """Organise la structure des dossiers COCO"""
        print("üóÇÔ∏è  Organisation de la structure COCO...")
        
        # Structure attendue apr√®s extraction:
        # coco/
        #   ‚îú‚îÄ‚îÄ train2017/
        #   ‚îú‚îÄ‚îÄ val2017/
        #   ‚îú‚îÄ‚îÄ test2017/
        #   ‚îî‚îÄ‚îÄ annotations/
        
        expected_dirs = ["train2017", "val2017", "annotations"]
        
        for dir_name in expected_dirs:
            dir_path = self.base_dir / dir_name
            if dir_path.exists():
                print(f"‚úì {dir_name} trouv√©")
            else:
                print(f"‚ö†Ô∏è  {dir_name} manquant")
        
        print("‚úì Structure organis√©e")

    def load_coco_annotations(self, split="train"):
        """Charge les annotations COCO"""
        print(f"üìã Chargement des annotations {split}...")
        
        ann_file = self.base_dir / "annotations" / f"instances_{split}2017.json"
        
        if not ann_file.exists():
            print(f"‚ùå Fichier d'annotations non trouv√©: {ann_file}")
            return None
        
        try:
            with open(ann_file, 'r') as f:
                coco_data = json.load(f)
            
            print(f"‚úì Annotations {split} charg√©es:")
            print(f"   üì∏ Images: {len(coco_data['images'])}")
            print(f"   üè∑Ô∏è  Annotations: {len(coco_data['annotations'])}")
            print(f"   üìÇ Cat√©gories: {len(coco_data['categories'])}")
            
            return coco_data
            
        except Exception as e:
            print(f"‚ùå Erreur lors du chargement: {e}")
            return None

    def create_class_folders(self, split="train"):
        """Cr√©e des dossiers par classe et organise les images"""
        print(f"üóÇÔ∏è  Cr√©ation des dossiers par classe pour {split}...")
        
        # Charger les annotations
        coco_data = self.load_coco_annotations(split)
        if not coco_data:
            return False
        
        # Cr√©er le dossier de sortie organis√©
        organized_dir = self.base_dir / f"{split}_organized"
        organized_dir.mkdir(exist_ok=True)
        
        # Cr√©er un mapping des cat√©gories
        categories = {cat['id']: cat['name'] for cat in coco_data['categories']}
        
        # Cr√©er les dossiers de classes
        for cat_id, cat_name in categories.items():
            class_dir = organized_dir / f"{cat_id:03d}_{cat_name.replace(' ', '_').replace('/', '_')}"
            class_dir.mkdir(exist_ok=True)
        
        # Cr√©er un mapping image_id -> annotations
        img_to_anns = {}
        for ann in coco_data['annotations']:
            img_id = ann['image_id']
            if img_id not in img_to_anns:
                img_to_anns[img_id] = []
            img_to_anns[img_id].append(ann)
        
        # Copier les images dans les bons dossiers
        images_dir = self.base_dir / f"{split}2017"
        
        print("üìÅ Organisation des images par classe...")
        for img_info in tqdm(coco_data['images'], desc="Images"):
            img_id = img_info['id']
            img_filename = img_info['file_name']
            img_path = images_dir / img_filename
            
            if not img_path.exists():
                continue
            
            # Obtenir les cat√©gories de cette image
            if img_id in img_to_anns:
                for ann in img_to_anns[img_id]:
                    cat_id = ann['category_id']
                    cat_name = categories[cat_id]
                    
                    # Dossier de destination
                    dest_dir = organized_dir / f"{cat_id:03d}_{cat_name.replace(' ', '_').replace('/', '_')}"
                    dest_path = dest_dir / img_filename
                    
                    # Copier l'image (si pas d√©j√† fait)
                    if not dest_path.exists():
                        shutil.copy2(img_path, dest_path)
        
        print(f"‚úì Images organis√©es dans {organized_dir}")
        return True

    def download_coco_captions(self):
        """T√©l√©charge les annotations de l√©gendes COCO"""
        caption_url = "http://images.cocodataset.org/annotations/annotations_trainval2017.zip"
        
        print("üí¨ T√©l√©chargement des l√©gendes COCO...")
        
        # Les l√©gendes sont incluses dans le fichier d'annotations principal
        ann_file = self.base_dir / "annotations" / "captions_train2017.json"
        
        if ann_file.exists():
            print("‚úì L√©gendes d√©j√† disponibles")
            
            # Charger et afficher des stats
            with open(ann_file, 'r') as f:
                captions_data = json.load(f)
            
            print(f"üìä Statistiques des l√©gendes:")
            print(f"   üì∏ Images: {len(captions_data['images'])}")
            print(f"   üí¨ L√©gendes: {len(captions_data['annotations'])}")
            
            return True
        else:
            print("‚ö†Ô∏è  T√©l√©chargez d'abord les annotations principales")
            return False

    def verify_coco_dataset(self):
        """V√©rifie l'int√©grit√© du dataset COCO"""
        print("üîç V√©rification du dataset COCO...")
        
        # V√©rifier les dossiers principaux
        required_dirs = ["train2017", "val2017", "annotations"]
        missing_dirs = []
        
        for dir_name in required_dirs:
            dir_path = self.base_dir / dir_name
            if not dir_path.exists():
                missing_dirs.append(dir_name)
        
        if missing_dirs:
            print(f"‚ùå Dossiers manquants: {', '.join(missing_dirs)}")
            return False
        
        # Compter les images
        stats = {}
        for split in ["train", "val"]:
            img_dir = self.base_dir / f"{split}2017"
            if img_dir.exists():
                images = list(img_dir.glob("*.jpg"))
                stats[split] = len(images)
                print(f"üìä {split}2017: {len(images)} images")
        
        # V√©rifier les annotations
        ann_dir = self.base_dir / "annotations"
        ann_files = list(ann_dir.glob("*.json"))
        print(f"üìã Fichiers d'annotations: {len(ann_files)}")
        
        for ann_file in ann_files:
            print(f"   üìÑ {ann_file.name}")
        
        print("‚úì V√©rification termin√©e")
        return True

    def create_coco_info(self):
        """Cr√©e un fichier d'informations sur COCO"""
        info = {
            "dataset_name": "MS COCO 2017",
            "description": "Microsoft Common Objects in Context - Dataset de d√©tection d'objets et de l√©gendes",
            "splits": {
                "train": "118,287 images",
                "val": "5,000 images", 
                "test": "40,670 images"
            },
            "categories": 80,
            "total_annotations": "~860,000 object instances",
            "captions": "~400,000 image captions",
            "image_format": "JPEG",
            "tasks": [
                "Object Detection",
                "Instance Segmentation", 
                "Image Captioning",
                "Panoptic Segmentation"
            ],
            "organization": {
                "images": "{split}2017/{image_id:012d}.jpg",
                "annotations": "annotations/{task}_{split}2017.json"
            }, 
            "source": "https://cocodataset.org/",
            "paper": "Microsoft COCO: Common Objects in Context (Lin et al., 2014)",
            "license": "Creative Commons Attribution 4.0",
            "download_size": "~25 GB total"
        }
        
        with open("coco_dataset_info.json", "w") as f:
            json.dump(info, f, indent=2, ensure_ascii=False)
        
        print("‚úì Informations sauvegard√©es dans coco_dataset_info.json")

    def download_coco_subset(self, max_images_per_class=100):
        """T√©l√©charge un sous-ensemble de COCO pour des tests rapides"""
        print(f"üéØ Cr√©ation d'un sous-ensemble COCO ({max_images_per_class} images/classe max)")
        
        # D'abord t√©l√©charger les annotations
        if not self.download_coco_dataset(["annotations"]):
            print("‚ùå Impossible de t√©l√©charger les annotations")
            return False
        
        self.extract_coco_files(["annotations"])
        
        # Charger les annotations
        coco_data = self.load_coco_annotations("val")  # Utiliser val (plus petit)
        if not coco_data:
            return False
        
        # Cr√©er le dossier subset
        subset_dir = self.base_dir / "coco_subset"
        subset_dir.mkdir(exist_ok=True)
        
        # S√©lectionner des images par cat√©gorie
        categories = {cat['id']: cat['name'] for cat in coco_data['categories']}
        
        # Grouper les annotations par cat√©gorie
        cat_to_images = {}
        for ann in coco_data['annotations']:
            cat_id = ann['category_id']
            img_id = ann['image_id']
            
            if cat_id not in cat_to_images:
                cat_to_images[cat_id] = set()
            cat_to_images[cat_id].add(img_id)
        
        # S√©lectionner les images √† t√©l√©charger
        selected_images = set()
        for cat_id, img_ids in cat_to_images.items():
            img_list = list(img_ids)[:max_images_per_class]
            selected_images.update(img_list)
        
        print(f"üìä Images s√©lectionn√©es: {len(selected_images)}")
        
        # T√©l√©charger les images s√©lectionn√©es
        # Note: Ici on devrait impl√©menter le t√©l√©chargement individuel
        # Pour l'instant, on recommande de t√©l√©charger val2017 complet
        print("üí° Pour un sous-ensemble, t√©l√©chargez val2017 complet puis filtrez")
        
        return True

def main():
    """Fonction principale avec menu interactif"""
    downloader = COCODownloader()
    
    print("üñºÔ∏è  === T√âL√âCHARGEUR MS COCO 2017 ===\n")
    print("MS COCO 2017 contient:")
    print("- Training: 118,287 images (~18 GB)")
    print("- Validation: 5,000 images (~1 GB)")
    print("- Test: 40,670 images (~6 GB)")
    print("- Annotations: 80 cat√©gories d'objets")
    print("- Taille totale: ~25 GB")
    print("- Licence: Creative Commons (libre)")
    print()
    
    while True:
        print("Options de t√©l√©chargement:")
        print("1. üì• T√©l√©charger validation set (recommand√© pour d√©buter)")
        print("2. üì• T√©l√©charger training set")
        print("3. üì• T√©l√©charger dataset complet")
        print("4. üì• T√©l√©charger annotations seulement")
        print("5. üì¶ Extraire fichiers t√©l√©charg√©s")
        print("6. üóÇÔ∏è  Organiser par classes")
        print("7. üîç V√©rifier dataset")
        print("8. ‚ÑπÔ∏è  Cr√©er fichier d'informations")
        print("9. üéØ T√©l√©charger sous-ensemble (test)")
        print("10. ‚ùå Quitter")
        
        choice = input("\nChoisissez une option (1-10): ").strip()
        
        if choice == "1":
            print("\n=== T√âL√âCHARGEMENT VALIDATION SET ===")
            if downloader.download_coco_dataset(["val_images", "annotations"]):
                downloader.extract_coco_files(["val_images", "annotations"])
                downloader.organize_coco_structure()
        
        elif choice == "2":
            print("\n=== T√âL√âCHARGEMENT TRAINING SET ===")
            print("‚ö†Ô∏è  Attention: 18 GB √† t√©l√©charger!")
            confirm = input("Continuer? (y/N): ")
            if confirm.lower() == 'y':
                if downloader.download_coco_dataset(["train_images", "annotations"]):
                    downloader.extract_coco_files(["train_images", "annotations"])
                    downloader.organize_coco_structure()
        
        elif choice == "3":
            print("\n=== T√âL√âCHARGEMENT COMPLET ===")
            print("‚ö†Ô∏è  Attention: ~25 GB √† t√©l√©charger!")
            confirm = input("Continuer? (y/N): ")
            if confirm.lower() == 'y':
                components = ["train_images", "val_images", "test_images", "annotations"]
                if downloader.download_coco_dataset(components):
                    downloader.extract_coco_files(components)
                    downloader.organize_coco_structure()
        
        elif choice == "4":
            print("\n=== T√âL√âCHARGEMENT ANNOTATIONS ===")
            if downloader.download_coco_dataset(["annotations"]):
                downloader.extract_coco_files(["annotations"])
        
        elif choice == "5":
            print("\n=== EXTRACTION FICHIERS ===")
            components = input("Composants √† extraire (train_images,val_images,annotations): ").strip()
            if components:
                comp_list = [c.strip() for c in components.split(",")]
                downloader.extract_coco_files(comp_list)
            else:
                downloader.extract_coco_files()
        
        elif choice == "6":
            print("\n=== ORGANISATION PAR CLASSES ===")
            split = input("Split √† organiser (train/val): ").strip() or "val"
            downloader.create_class_folders(split)
        
        elif choice == "7":
            print("\n=== V√âRIFICATION DATASET ===")
            downloader.verify_coco_dataset()
        
        elif choice == "8":
            print("\n=== CR√âATION INFOS ===")
            downloader.create_coco_info()
        
        elif choice == "9":
            print("\n=== SOUS-ENSEMBLE TEST ===")
            max_images = int(input("Images max par classe (d√©faut 50): ") or "50")
            downloader.download_coco_subset(max_images)
        
        elif choice == "10":
            print("Au revoir!")
            break
        
        else:
            print("Option invalide!")
        
        print("\n" + "="*50 + "\n")

if __name__ == "__main__":
    main()
```
